```python
import numpy as np
import plotly.graph_objects as go
```

```python

n = 100
x = np.random.normal(size=(1000,))
y = np.sin(x) + np.random.random(size=x.shape)

x[:5], y[:5]
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x,
    y=y,
    marker=dict(
        color=y,
        colorscale='oranges'
    ),
    mode='markers'
))

fig.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.cm import summer

fig, ax = plt.subplots()

ax.scatter(x=x, y=y, c=y, cmap=summer);
```
